```python
#Seaborn and matplotlib combinly

#tips dataset also available in default in python programming
import seaborn as sns
tips =  sns.load_dataset('tips')
tips
```

```python
# we are creating bar graphs for iris

plt.bar(data['species'],data['petal_length'])
plt.bar(data['species'],data['petal_width'])

plt.show()
```

```python
plt.bar(tips["day"],tips['total_bill'])
```

```python
plt.barh(tips["day"],tips['total_bill'])
```

```python
# Histogram 
plt.hist(tips['total_bill'], edgecolor= "black")
plt.show()
```

```python
#Day must be in fixed length

day = tips.groupby('day')['total_bill'].sum().reset_index()

day
```

```python
# Pie Chart

x= tips['day']
y=tips['total_bill']
tips
```

```python
x
```

```python
# Day must be fixed length

day=tips.groupby('day')['total_bill'].sum().reset_index()

day
```

```python
# Pie Chart

x= day['day']
y=day['total_bill']
```

```python
explode = [0,0,0,0.1]

plt.pie(y,labels=x,explode = explode,autopct='%1.1f%%',colors=['red','blue','green','yellow'],wedgeprops = {'edgecolor' : 'black'},shadow=True,startangle=90)
plt.show()
```

```python
# scatter plot

plt.scatter(tips['total_bill'],tips['tip'])
plt.title("Scatter plot",fontsize=17)
plt.xlabel("Total Bill")
plt.ylabel("Tip")

plt.show()
```

```python
tips['size'].value_counts()
```

```python
# I wnat to create scatter plot with some color palet

# scatter plot

plt.scatter(tips['total_bill'],tips['tip'],s=100,c=tips['size'])
plt.title("Scatter plot",fontsize=17)
plt.xlabel("Total Bill")
plt.ylabel("Tip")
cbar=plt.colorbar()
cbar.set_label("Size")

plt.show()
```
